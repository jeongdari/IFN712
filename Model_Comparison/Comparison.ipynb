```python
import matplotlib.pyplot as plt
import numpy as np

# Data for each station and model improvement rates
stations = ['HKSL', 'HKLM', 'HKOH', 'HKKT']
models = ['Random Forest', 'FCNN', 'CNN', 'LSTM', 'Decision Tree', 'Linear Model', 'RBF SVR', 'Tuned CNN']

# Improvement rates for each model at each station
improvement_rates = {
    'HKSL': [43.88, 40.76, 41.28, 36.75, 15.58, 39.58, 38.68, 39.89],
    'HKLM': [38.16, 29.48, 31.96, 23.76, 8.45, 23.66, 29.14, 27.54],
    'HKOH': [35.22, 25.52, 28.43, 29.93, 5.91, 19.20, 33.07, 33.62],
    'HKKT': [27.36, 25.54, 30.69, 25.04, 0.09, 21.80, 31.69, 33.67]
}

# Define colors for the bins
colors_high = 'orange'  # Highest
colors_second = 'green'  # Second-highest
colors_third = 'purple'  # Third-highest
colors_default = 'skyblue'  # All others

# Create subplots for each station
fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharey=True)
fig.suptitle('Improvement Rates by Model and GNSS Station', fontsize=16)

# Plotting improvement rates for each station
for i, station in enumerate(stations):
    ax = axes[i//2, i%2]
    rates = improvement_rates[station]
    
    # Find the indices of highest, second-highest, and third-highest rates
    sorted_indices = np.argsort(rates)[::-1]
    color_mapping = [colors_high if idx == sorted_indices[0] 
                     else colors_second if idx == sorted_indices[1] 
                     else colors_third if idx == sorted_indices[2]
                     else colors_default for idx in range(len(rates))]
    
    # Plot the horizontal bar chart with color mapping
    ax.barh(models, rates, color=color_mapping)
    ax.set_title(f'{station} Station')
    ax.set_xlim(0, 45)  # Set x-limit to ensure uniform scale
    ax.set_xlabel('Improvement Rate (%)')

# Add a legend for the color codes
handles = [plt.Rectangle((0,0),1,1, color=colors_high), 
           plt.Rectangle((0,0),1,1, color=colors_second), 
           plt.Rectangle((0,0),1,1, color=colors_third), 
           plt.Rectangle((0,0),1,1, color=colors_default)]
labels = ['Highest Improvement', 'Second-Highest Improvement', 'Third-Highest Improvement', 'Others']

fig.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, 0), ncol=4, fontsize=12)

# Adjust layout and show plot
plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.show()
```
