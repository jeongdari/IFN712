```python
import matplotlib.pyplot as plt

# Replace these with the actual MAE values from your separate notebooks
cnn_mae = 0.10859262943267822  # From CNN_first.ipynb
dt_mae = 0.1287378332134963
fcnn_mae = 0.10655459468799028  # From FCNN.ipynb
random_forest_mae = 0.09623093139961929  # From RandomForest.ipynb
rnn_mae = 0.10765134026776775  # From RNN.ipynb
svm_mae = 0.11999214675255403  # From SVM.ipynb

# Dictionary of models and their MAE values
mae_values = {
    "CNN": cnn_mae,
    "Decision Tree": dt_mae,
    "FCNN": fcnn_mae,
    "Random Forest": random_forest_mae,
    "RNN": rnn_mae,
    "SVM": svm_mae
}

# Convert MAE values to lists
models = list(mae_values.keys())
mae_scores = list(mae_values.values())

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(models, mae_scores, color='skyblue')

# Add title and labels
plt.title('HKSL Model Performance Comparison (MAE)', fontsize=16)
plt.xlabel('Model', fontsize=14)
plt.ylabel('Mean Absolute Error (MAE)', fontsize=14)

# Display the bar chart
plt.show()
```
